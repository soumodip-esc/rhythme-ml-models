# Step 3: Training Model

```python
from sklearn.model_selection import train_test_split
```

```python
#Encoding 'frequency' Colums
df['frequency_encoded'] = df['frequency'].map({'daily': 0, 'weekly': 1, 'monthly': 2, 'twice_weekly': 3})
```

```python
df.columns
```

```python
# Select your input features (X) and target variable (y)
x = df[[
    'day_of_week',
    'is_weekend',
    'current_streak',
    'completion_rate_7d',
    'completion_rate_30d',
    'days_since_start',
    'frequency_encoded'
]]
y = df['is_completed']

print("Features shape:", x.shape)
print("Target shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train ,X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

print(f"Training set: {X_train.shape[0]} samples")
print(f"Testing set: {X_test.shape[0]} samples")
print(f"Total samples: {len(x)}")
```

```python
# ========== OPTIMIZED MODEL FOR YOUR FAKE DATA ==========

model = RandomForestClassifier(
    n_estimators=100,
    max_depth=8,
    min_samples_split=15,
    min_samples_leaf=8,
    max_features='sqrt',
    random_state=42,
    class_weight='balanced'
)

model.fit(X_train, y_train)

# Evaluate
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print(f"Training Accuracy: {train_score:.1%}")
print(f"Testing Accuracy:  {test_score:.1%}")
print(f"Overfitting Gap:   {(train_score - test_score)*100:.1f}%")

#  GOOD if gap is < 10%
#  CONCERNING if gap is > 15%
```

```python
test_data = [[3, 0, 15, 0.7, 0.6, 50, 1]]
prob_heavy = model.predict_proba(test_data)[0][1]
print(f"Heavy Regularization: {prob_heavy:.3f}")

```

# Step 4: Model Training and Evaluation

```python
# Train the model
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# Make predictions
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print(f"\nModel accuracy: {accuracy:.3f} ({accuracy*100:.1f}%)")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1 Score: {f1:.3f}")
```

```python
metrics = [accuracy, precision, recall, f1]
labels = ['Accuracy', 'Precision', 'Recall', 'F1']
plt.figure(figsize=(8, 5))
plt.bar(labels, metrics)
plt.title('Model Performance')
plt.ylim(0, 1)
plt.show()
```
